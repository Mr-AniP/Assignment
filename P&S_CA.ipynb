Q1 Answer

```python
import numpy as np
import matplotlib.pyplot as ax

# Generate 100 samples
data = np.random.uniform(0,1,100)

ax.rcParams["figure.figsize"] = [8, 5]

fig_histo=ax.figure("Histo fig")
histogram = ax.hist(data, 100)
ax.title("Histogram of distribution")

fig_histo=ax.figure("CDF fig")
pdf = histogram[0]/sum(histogram[0])
cdf = pdf.cumsum()

ax.plot(histogram[1][1:],cdf)
ax.title("CDF of distribution")
ax.show()
```

Q4 Answer

```python
import numpy as np
import matplotlib.pyplot as ax

ax.rcParams["figure.figsize"] = [8, 5]

# taking m & p as inputs
m=int(input())
p=float(input())

# setting up Geomretric RV
geometric=np.zeros(m)
bernouli=np.zeros(1,dtype=int)
for i in range(m):
  count=0
  while(bernouli[0]==0):
    count=count+1
    bernouli = np.where(np.random.uniform(0,1,1)< p, 1, 0)
  geometric[i]=count
  bernouli[0]=0

ax.figure("Geometric plot")
ax.title(("Histogram showing generation of "+ str(m) + " samples from a Geometric RV with paramater "+str(p)+" using Bernoulli RV"))
ax.hist(geometric,rwidth = 0.8,bins=np.arange(geometric.min()-0.5, geometric.max()+1.50))
ax.xticks(np.arange(geometric.min(),geometric.max()+1))
ax.xlabel('Value')
ax.ylabel('Frequency')
ax.show()
```

Q5 Answer

```python
import numpy as np
import matplotlib.pyplot as ax

ax.rcParams["figure.figsize"] = [8, 5]

# taking m, n & p as inputs
m=int(input())
n=int(input())
p=float(input())

# Generate m samples
count=0
binomial=np.zeros(m,dtype=int)
while(count!=m):
  bernoulli = np.where(np.random.uniform(0,1,n) < p, 1, 0) # setting up Bernoulli RV
  binomial[count]=bernoulli.sum()   # setting up Binomial RV
  count+=1

ax.figure("Binomial plot")
ax.title(("Histogram showing generation of "+ str(m) + " samples from a Binomial RV with paramaters ("+str(n)+","+str(p)+") using Bernoulli RV"))
ax.hist(binomial,rwidth = 0.8,bins=np.arange(binomial.min()-0.5, binomial.max()+1.50))
ax.xticks(np.arange(binomial.min(),binomial.max()+1))
ax.xlabel('Value')
ax.ylabel('Frequency')
ax.show()
```

```python
import numpy as np
import matplotlib.pyplot as ax
from scipy.stats import poisson as poisso



ax.rcParams["figure.figsize"] = [8, 5]

# taking m & alpha as inputs
m=int(input())
p=float(input())

# Generate m samples
poisson= poisso.rvs(p, size = m)


ax.figure("Poisson plot")
ax.title(("Histogram showing generation of "+ str(m) + " samples from a Poisson RV with paramater "+str(p)))
ax.hist(poisson,rwidth = 0.8,bins=np.arange(poisson.min()-0.5, poisson.max()+1.50))
ax.xticks(np.arange(poisson.min(),poisson.max()+1))
ax.xlabel('Value')
ax.ylabel('Frequency')
ax.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as ax

ax.rcParams["figure.figsize"] = [8, 5]

# taking m, mean & standard deviation as inputs
m=int(input())
mn=int(input())
std=float(input())

# Generate m samples
gaussian=np.zeros(m)
for i in range(m):
  gaussian[i]=random.gauss(mn,std)

ax.figure("Gaussian plot")
ax.title(("Histogram showing generation of "+ str(m) + " samples from a Gaussian RV with paramaters ("+str(mn)+","+str(std)+")"))
ax.hist(gaussian,rwidth = 0.8,bins=int(m/10))
ax.xlabel('Value')
ax.ylabel('Frequency')
ax.show()
```

Q9 Answer

```python
import numpy as np
import matplotlib.pyplot as ax

ax.rcParams["figure.figsize"] = [8, 5]

# taking p as input
p=float(input())
x=[]
y=[]
data= np.random.uniform(0,1,1000) # Generate m samples
bernoulli = np.where(data < p, 1, 0) # setting up Bernoulli RV
for m in range(1,1001):
  x.append(m)
  y.append((bernoulli[:m].sum()/m)) # Generating means

ax.figure("Mean plot")
ax.title(("Plot showing sample mean for each m"))
ax.plot(x,y)
ax.xlabel('Samples')
ax.ylabel('Sample mean')

ax.figure("P plot")
ax.title(("Plot showing p for each m"))
ax.plot(x,data)
ax.xlabel('Samples')
ax.ylabel('p')

ax.show()
```

Yes we see that sample mean converge to the expected value of Bernoulli RV.

The smallest value of m for which the sample mean has more or less converged to the parameter of the Bernoulli RV is '600'
