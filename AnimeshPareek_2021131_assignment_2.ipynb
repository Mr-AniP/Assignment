````python
def fit_linearegressionmodel(x,y,s):
  b0 = np.random.rand()
  b1 = np.random.rand()
  # print(b0,b1)
  i=0
  while(True):
    if(i>=500):
      break
    y_pred=b0 +b1 *x
    # print(b0,b1)
    error = y_pred-y
    b0 -= s* np.sum(error)
    b1 -= s* np.sum(error * x)
    i+=1
  print(np.mean(error))
  return b0,b1
````

````python
# g=np.stack(y).T
X = np.stack((y,np.ones((x1.shape[0])),x1,x2,x3))
def mypca(X,size):
  X_std=((X - np.mean(X, axis=0)) / np.std(X, axis=0))
  eval, evect = np.linalg.eig(np.cov(X_std, rowvar=False))
  ind = np.argsort(eval) [::-1]
  eval=np.take_along_axis(eval,ind,axis=0)
  evect=evect[:,ind]
  return np.dot(X_std,evect[:, :size])
X=mypca(X,100)
g=X[0,:]
X=X[1:,:]
cef=np.dot(np.linalg.inv(np.dot(X,X.T)),np.dot(X,g))
print("MLRM after pca")
print(cef)
print( "order -> b_intercept , b_education, b_occupation, b_experiance")
mean_error=np.mean((g-np.dot(X.T,cef))**2)
print("mean of the error =",mean_error)
print()
````

Observation ->


```
Clearly mean error is very less then both Polynomial regression and MLRM model.

This is because we left out some details (with lessor varience) as a result we see this change..

```

````python
def orthog(data):
  orthogonalized_data = np.zeros_like(data, dtype=float)
  for i in range(data.shape[1]):
    orthogonalized_data[:, i]=data[:, i] - sum([(np.dot(data[:, i], orthogonalized_data[:, j]) / np.dot(orthogonalized_data[:, j], orthogonalized_data[:, j])) * orthogonalized_data[:, j] for j in range(i)])
  return orthogonalized_data
X = np.stack((np.ones((x1.shape[0])),x1,x2,x3))
mat=np.cov(X)
g=np.stack(y).T
if(all(abs(mat[i][j])<1e-9 for i in range(len(mat)) for j in range(len(mat[i])) if(i!=j))):
  print("Data is orthogonal !!")
else:
  print("Data is not orthogonal !!")
  X=orthog(X.T)
  X=X.T
cef=np.dot(np.linalg.inv(np.dot(X,X.T)),np.dot(X,g))
print("MLRM_orthogonalized")
print(cef)
print( "order -> b_intercept , b_education, b_occupation, b_experiance")
mean_error=np.mean((g-np.dot(X.T,cef))**2)
print("mean of the error =",mean_error)
print()
````

Observation ->


```
After orthogonalization our model gives more or less same mean square error as compared to  MLRM model.
```
